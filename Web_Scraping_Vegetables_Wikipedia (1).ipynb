# Importing libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# Making a request

```python
url = "https://en.wikipedia.org/wiki/Vegetable"
req = requests.get(url)
```

# Extracting contents

```python
soup = BeautifulSoup(req.text,'html.parser')
soup
```

<b>OBJECTIVES: <b><br>

1) Create a dataframe - Top producers

2) Create a dataframe - Some common vegetables

# Creating Top Producers dataframe

```python
table1 = soup.find('table',class_ = "sortable wikitable")
print(table1)
```

Step 2 : Extracting text from the table1

```python
#Extracting headings from the table
th = table1.find_all('th')
print(th)
```

```python
#Extracting text from th
columns = [i.text.strip() for i in th]
print(columns)
```

```python
#Extracting all trs'  from the table
tr = table1.find_all('tr')
print(tr)
```

```python
#Extracting all text from trs
tds = [i.text.strip() for i in tr]
print(tds)
```

```python
table_contents = [i.split('\n') for i in tds]
```

```python
#pop the 1st item from table_contents
table_contents.pop(0)
```

```python
print(table_contents)
```

Step 3: Creating a dataframe

```python
top_producers = pd.DataFrame(table_contents,columns = columns)
top_producers
```

# Creating Some Common Vegetables dataframe

```python
table2 = soup.find('table',class_ = "wikitable sortable")
table2
```

```python
#Extracting headings from table2
th2 = table2.find_all('th')
print(th2)
```

```python
table2_headings = [i.text.strip() for i in th2]
print(table2_headings)
```

```python
columns2 = table2_headings
columns2.remove('Image')
```

```python
#Extracting rows from table2
tr2 = table2.find_all('tr')
print(tr2)
```

```python
tdb = [i.text.strip() for i in tr2]
tdb
```

```python
table2_content = [i.split('\n') for i in tdb]
table2_content = table2_content[2:]
table2_content
```

```python
table2_contents = []
for i in table2_content:
  k = [j.strip() for j in i]
  table2_contents.append(k)


print(table2_contents)
```

```python
#creating a dataframe

table2 = pd.DataFrame(table2_contents,columns = columns2)
table2
```
